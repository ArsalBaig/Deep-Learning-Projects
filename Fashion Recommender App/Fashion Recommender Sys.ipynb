```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import os

from tqdm import tqdm
from PIL import Image # Use for dealing with open, close, & manipulation of images.
from numpy.linalg import norm
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing import image
from tensorflow.keras.layers import GlobalMaxPool2D
from sklearn.neighbors import NearestNeighbors
from tensorflow.keras.applications.resnet50 import ResNet50, preprocess_input
```

# ResNet50 Model

```python
RestNet50_model = ResNet50(weights= 'imagenet', include_top= False, input_shape= (224, 224, 3))
RestNet50_model.trainable = False
RestNet50_model = Sequential([RestNet50_model, GlobalMaxPool2D()])
```

# Feature Extraction Func

```python
def extract_features(image_path, model):
    loaded_image = image.load_img(image_path, target_size= (224, 224))
    # Convert img to Array and expand dim's.
    image_array = image.img_to_array(loaded_image)
    expand_dim = np.expand_dims(image_array, axis= 0)
    # Preprocess Image.
    preprocess_image = preprocess_input(expand_dim)
    # Predict & flatten the image.
    model_pred = model.predict(preprocess_image)
    flatten_image = model_pred.flatten()
    model_norm = norm(flatten_image)
    return model_norm
```

# Listing image file & features

```python
images_file = []
for fashion_imgs in os.listdir('/content/myntradataset/images')[:3000]:
    images_path = os.path.join('/content/myntradataset/images', fashion_imgs) # combines directory path & filename.
    images_file.append(images_path)
```

```python
images_features = []
for file in tqdm(images_file):
    features = extract_features(file, RestNet50_model)
    images_features.append(features)
```

# Saving Model.

```python
import pickle

pickle.dump(images_file, open('images_file.pkl', 'wb'))
pickle.dump(images_features, open('images_features.pkl', 'wb'))
```
